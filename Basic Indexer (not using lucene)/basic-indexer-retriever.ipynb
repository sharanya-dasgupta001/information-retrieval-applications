```python
import pandas as pd
import nltk
from nltk import word_tokenize
porter = nltk.PorterStemmer()
```

```python
file_path = 'D:\OneDrive\Desktop\I S I\Information Retrieval (Debapriyo Majumdar)\PyLucene\wiki_movie_plots_deduped.csv'
```

# INDEXER

```python
def make_inverted_index(file_path) :
    df = pd.read_csv(file_path)
    df = df['Plot']
    # A dictionary with key as the terms and the values are the lists containing document frequency of 
    # the term and posting list
    inverted_index = {}
    doc_id = 0
    # Processing the content and making inverted index
    for movie in df :
        print("Movie -",doc_id)
        # Tokenize plots
        tokens = word_tokenize(movie)
        # Convert to lowercase and remove punctuations
        tokens_lower = [w.lower() for w in tokens if w.isalnum()]
        #Stemming
        stemmed = [porter.stem(w) for w in tokens_lower]
        for stem in stemmed:
            val = inverted_index.get(stem) # if key is not present [] -> get error , .get() -> None 
            if val == None :
                ls = [1,[doc_id]]
                inverted_index[stem] = ls
            else :
                ls = inverted_index[stem]
                # ls[0] -> document frequency
                # ls[1] -> doc_id
                if doc_id not in ls[1]:
                    ls[1].append(doc_id)
                    ls[0]+=1
        doc_id += 1
    return inverted_index
```

```python
inverted_index = make_inverted_index(file_path)
```

# Pickle

```python
import pickle
def DumpData(d,f):
    fd = open(f,'ab')
    pickle.dump(d,fd)
    fd.close()
def loadData(pickle_file):
    fd = open(pickle_file,'rb')
    d = pickle.load(fd)
    fd.close()
    return d
```

```python
DumpData(inverted_index,'inverted_dump.pickle')
```

```python
inverted = loadData('inverted_dump.pickle')
```

# RETRIEVER

```python
def processText(text):
    # Tokenize plots
    tokens = word_tokenize(text)
    # Convert to lowercase and remove punctuations
    tokens_lower = [w.lower() for w in tokens if w.isalnum()]
    #Stemming
    stemmed = [porter.stem(w) for w in tokens_lower]
    return stemmed
```

```python
Q = input("Enter the query :")
Q = processText(Q)
print(Q)
```

```python
all_pl = []
for q in Q:
    pl = inverted.get(q)
    if pl is not None:
        all_pl.append(pl[1])
print(all_pl)
```
